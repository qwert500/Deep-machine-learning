# Create the structure (Flower Recognition)

This notebook creates the folder structure necessary for the project using the Flowers Recognition
dataset from Kaggle. This should be run from the same folder where the `flowers-recognition.zip` file is. This notebook is a modified copy from the HA1 assignment.

```python
# For dealing with files
import os
import shutil

# For using regex expressions
import re

# For splitting the data
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# save home path (for coming back afterwards)
home_path = "/Users/timhell/Google Drive/11 FaÌˆrdigheter/Programmering/Python/Deep machine learning/deep-machine-learning/Project"
#home_path = "/home/student/deep-machine-learning/Project"
flowers_path = home_path + '/flowers/'
daisy_path = flowers_path + '/daisy/'
dandelion_path = flowers_path + '/dandelion/'
rose_path = flowers_path + '/rose/'
sunflower_path = flowers_path + '/sunflower/'
tulip_path = flowers_path + '/tulip/'

# cd to home directory
os.chdir(home_path)

if os.path.exists('flowers'):
    shutil.rmtree('flowers')
if os.path.exists('train'):
    shutil.rmtree('train')
if os.path.exists('val'):
    shutil.rmtree('val')
if os.path.exists('test'):
    shutil.rmtree('test')

# Depending on your machine the following might take some seconds to run
!unzip -q flowers-recognition.zip
```

```python
# Create the train directories
if not os.path.isdir('train'):
    os.mkdir('train')
    
if not os.path.isdir('train/daisy'):
    os.mkdir('train/daisy')
    
if not os.path.isdir('train/dandelion'):
    os.mkdir('train/dandelion')
    
if not os.path.isdir('train/rose'):
    os.mkdir('train/rose')
    
if not os.path.isdir('train/sunflower'):
    os.mkdir('train/sunflower')
    
if not os.path.isdir('train/tulip'):
    os.mkdir('train/tulip')

# Create the validation directories
if not os.path.isdir('val'):
    os.mkdir('val')
    
if not os.path.isdir('val/daisy'):
    os.mkdir('val/daisy')
    
if not os.path.isdir('val/dandelion'):
    os.mkdir('val/dandelion')
    
if not os.path.isdir('val/rose'):
    os.mkdir('val/rose')
    
if not os.path.isdir('val/sunflower'):
    os.mkdir('val/sunflower')
    
if not os.path.isdir('val/tulip'):
    os.mkdir('val/tulip') 
    
# Create the test directories    
if not os.path.isdir('test'):
    os.mkdir('test')
    
if not os.path.isdir('test/daisy'):
    os.mkdir('test/daisy')
    
if not os.path.isdir('test/dandelion'):
    os.mkdir('test/dandelion')
    
if not os.path.isdir('test/rose'):
    os.mkdir('test/rose')
    
if not os.path.isdir('test/sunflower'):
    os.mkdir('test/sunflower')
    
if not os.path.isdir('test/tulip'):
    os.mkdir('test/tulip')    
    
```

```python
# Get a list of all filenames (these will be used for training, validation and testing)
os.chdir(daisy_path)
all_daisy_filenames = np.array(os.listdir())
os.chdir(dandelion_path)
all_dandelion_filenames = np.array(os.listdir())
os.chdir(rose_path)
all_rose_filenames = np.array(os.listdir())
os.chdir(sunflower_path)
all_sunflower_filenames = np.array(os.listdir())
os.chdir(tulip_path)
all_tulip_filenames = np.array(os.listdir())
```

```python
print(all_rose_filenames[65])
```

```python
# Split it into training, validation and test sets
seed = 0
split_ratio_val = 0.0
split_ratio_test = 0.1
split_ratio_train = 1-split_ratio_val-split_ratio_test

# Split daisy
n_daisy = len(all_daisy_filenames)
np.random.seed(seed)
daisy_index = np.random.choice(3, n_daisy, p=[split_ratio_train, split_ratio_val, split_ratio_test])
daisy_train_mask = daisy_index==0
daisy_val_mask = daisy_index==1
daisy_test_mask = daisy_index==2
daisy_train_filenames = all_daisy_filenames[daisy_train_mask]
daisy_val_filenames = all_daisy_filenames[daisy_val_mask]
daisy_test_filenames = all_daisy_filenames[daisy_test_mask]

# Split dandelion
n_dandelion = len(all_dandelion_filenames)
np.random.seed(seed)
dandelion_index = np.random.choice(3, n_dandelion, p=[split_ratio_train, split_ratio_val, split_ratio_test])
dandelion_train_mask = dandelion_index==0
dandelion_val_mask = dandelion_index==1
dandelion_test_mask = dandelion_index==2
dandelion_train_filenames = all_dandelion_filenames[dandelion_train_mask]
dandelion_val_filenames = all_dandelion_filenames[dandelion_val_mask]
dandelion_test_filenames = all_dandelion_filenames[dandelion_test_mask]

# Split rose
n_rose = len(all_rose_filenames)
np.random.seed(seed)
rose_index = np.random.choice(3, n_rose, p=[split_ratio_train, split_ratio_val, split_ratio_test])
rose_train_mask = rose_index==0
rose_val_mask = rose_index==1
rose_test_mask = rose_index==2
rose_train_filenames = all_rose_filenames[rose_train_mask]
rose_val_filenames = all_rose_filenames[rose_val_mask]
rose_test_filenames = all_rose_filenames[rose_test_mask]

# Split sunflower
n_sunflower = len(all_sunflower_filenames)
np.random.seed(seed)
sunflower_index = np.random.choice(3, n_sunflower, p=[split_ratio_train, split_ratio_val, split_ratio_test])
sunflower_train_mask = sunflower_index==0
sunflower_val_mask = sunflower_index==1
sunflower_test_mask = sunflower_index==2
sunflower_train_filenames = all_sunflower_filenames[sunflower_train_mask]
sunflower_val_filenames = all_sunflower_filenames[sunflower_val_mask]
sunflower_test_filenames = all_sunflower_filenames[sunflower_test_mask]

# Split tulip
n_tulip = len(all_tulip_filenames)
np.random.seed(seed)
tulip_index = np.random.choice(3, n_tulip, p=[split_ratio_train, split_ratio_val, split_ratio_test])
tulip_train_mask = tulip_index==0
tulip_val_mask = tulip_index==1
tulip_test_mask = tulip_index==2
tulip_train_filenames = all_tulip_filenames[tulip_train_mask]
tulip_val_filenames = all_tulip_filenames[tulip_val_mask]
tulip_test_filenames = all_tulip_filenames[tulip_test_mask]
```

```python
# Move images to training directory
os.chdir(home_path)
for f in daisy_train_filenames:
    shutil.copyfile(daisy_path + f,'./train/daisy/' + f)
    
for f in dandelion_train_filenames:
    shutil.copyfile(dandelion_path + f,'./train/dandelion/' + f)   
    
for f in rose_train_filenames:
    shutil.copyfile(rose_path + f,'./train/rose/' + f)
    
for f in sunflower_train_filenames:
    shutil.copyfile(sunflower_path + f,'./train/sunflower/' + f)  
    
for f in tulip_train_filenames:
    shutil.copyfile(tulip_path + f,'./train/tulip/' + f)  
    
# Move images to validation directory
for f in daisy_val_filenames:
    shutil.copyfile(daisy_path + f,'./val/daisy/' + f)
    
for f in dandelion_val_filenames:
    shutil.copyfile(dandelion_path + f,'./val/dandelion/' + f)   
    
for f in rose_val_filenames:
    shutil.copyfile(rose_path + f,'./val/rose/' + f)
    
for f in sunflower_val_filenames:
    shutil.copyfile(sunflower_path + f,'./val/sunflower/' + f)  
    
for f in tulip_val_filenames:
    shutil.copyfile(tulip_path + f,'./val/tulip/' + f) 
    
# Move images to testing directory
for f in daisy_test_filenames:
    shutil.copyfile(daisy_path + f,'./test/daisy/' + f)
    
for f in dandelion_test_filenames:
    shutil.copyfile(dandelion_path + f,'./test/dandelion/' + f)   
    
for f in rose_test_filenames:
    shutil.copyfile(rose_path + f,'./test/rose/'+f)
    
for f in sunflower_test_filenames:
    shutil.copyfile(sunflower_path + f,'./test/sunflower/' + f)  
    
for f in tulip_test_filenames:
    shutil.copyfile(tulip_path + f,'./test/tulip/' + f)     
```
